```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype == 'object':
            df[col] = df[col].fillna(df[col].mode()[0])
        # If the column is numerical, fill missing values with the median.
        else:
            df[col] = df[col].fillna(df[col].median())
```

```python
df_null_counts = df.isnull().sum()
```

```python
def get_grade(score):
    if score >= 85: return 'A'
    elif score >= 75: return 'B'
    elif score >= 65: return 'C'
    elif score >= 55: return 'D'
    else: return 'F'

df['Grade'] = df['Exam_Score'].apply(get_grade)
```

```python
X = df.drop(['Exam_Score', 'Grade'], axis=1)
y = df['Grade']
```

```python
categorical_cols = X.select_dtypes(include=['object']).columns
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    label_encoders[col] = le
```

```python
y_le = LabelEncoder()
y_encoded = y_le.fit_transform(y)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"\n Train set: {X_train.shape}, Test set: {X_test.shape}")
print(f" Target classes: {y_le.classes_}")
```

```python

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
import lightgbm as lgb
lgb_model = lgb.LGBMClassifier(
    n_estimators=300,
    max_depth=7,
    learning_rate=0.05,
    num_leaves=50,
    random_state=42
)
```

```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)
```

```python
rf_model.fit(X_train, y_train)
```

```python
print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_model.predict(X_test)):.4f}")
```

```python

import pandas as pd
import numpy as np
def predict_student_grade(*args):
    input_df = pd.DataFrame([args], columns=X.columns)
    for col in categorical_cols:
        # Convert input to string to ensure it matches the encoder's training data
        input_df[col] = label_encoders[col].transform(input_df[col].astype(str))

    input_scaled = scaler.transform(input_df)

    # 4. Get probabilities from the Random Forest model only
    # predict_proba returns an array of probabilities for each class
    probabilities = rf_model.predict_proba(input_scaled)[0]

    # 5. Find the index of the highest probability
    predicted_index = np.argmax(probabilities)
    confidence_value = probabilities[predicted_index]

    # 6. Decode the numerical index back to the Letter Grade (A, B, C, etc.)
    predicted_grade = y_le.inverse_transform([predicted_index])[0]

    # Return the grade and the confidence formatted as a percentage
    return predicted_grade, f"{confidence_value:.2%}"

# Create input components dynamically based on your features
inputs = []
for col in X.columns:
    if col in categorical_cols:
        # Categorical columns become Dropdowns
        inputs.append(gr.Dropdown(
            choices=list(label_encoders[col].classes_),
            label=col.replace('_', ' ')
        ))
    else:
        # Numerical columns become number inputs with the mean as default
        inputs.append(gr.Number(
            label=col.replace('_', ' '),
            value=float(df[col].mean())
        ))

# Set up the UI
demo = gr.Interface(
    fn=predict_student_grade,
    inputs=inputs,
    outputs=[
        gr.Textbox(label="Predicted Grade"),
        gr.Textbox(label="Model Confidence")
    ],
    title="🎓 Student Performance Predictor (Random Forest)",
    description="Enter student data to predict the final grade. This model uses Random Forest classification.",
    theme="soft"
)

# Launch the app
if __name__ == "__main__":
    demo.launch(share=True)
```
